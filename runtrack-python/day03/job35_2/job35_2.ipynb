```python
import re
import matplotlib.pylab as plt
```

```python
f = open('../job33/data.txt','r')
txt = f.read()
```

```python
words = re.findall('\w+',txt)
dic = {}
for w in words:
    if len(w) in dic:
        dic[len(w)] += 1
    else:
        dic[len(w)] = 1

dic = dict(sorted(dic.items()))
occurencies = list(dic.values())
plt.bar(range(1,len(occurencies)+1), occurencies)
plt.show()
```
