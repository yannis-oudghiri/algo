```python
import re
import matplotlib.pylab as plt
```

```python
f = open('../job33/data.txt','r')
txt = f.read()
```

```python
txt_lower = txt.lower()
first_letter = re.findall('(?:\W|^)(\w)',txt_lower)
alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
occurencies = []
for letter in alphabet:
    occurencies.append(first_letter.count(letter))
plt.bar(range(len(occurencies)), occurencies, tick_label=alphabet)
plt.show()
```
